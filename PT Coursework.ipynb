```python
#compute mean return
BNP_mean = df_return['BNP Return'].mean()
TSLA_mean = df_return['TSLA Return'].mean()
PRAD_mean = df_return['PRAD Return'].mean()
BX_mean = df_return['BX Return'].mean()
RYA_mean = df_return['RYA Return'].mean()
AMZN_mean = df_return['AMZN Return'].mean()

#compute standard deviations
BNP_std = df_return['BNP Return'].std()
TSLA_std = df_return['TSLA Return'].std()
PRAD_std = df_return['PRAD Return'].std()
BX_std = df_return['BX Return'].std()
RYA_std = df_return['RYA Return'].std()
AMZN_std = df_return['AMZN Return'].std()


#Compute sharpe ratio
rf = 0.001
BNP_sharpe = (BNP_mean - rf)/BNP_std
TSLA_sharpe = (TSLA_mean - rf)/TSLA_std
PRAD_sharpe = (PRAD_mean - rf)/PRAD_std
BX_sharpe = (BX_mean - rf)/BX_std
RYA_sharpe = (RYA_mean - rf)/RYA_std
AMZN_sharpe = (AMZN_mean - rf)/AMZN_std
```

```python
mySeq = np.arange(0,101,2)
mySeq
```

```python
myCount = 0

for i in mySeq:
    tmp = 100 - i
    for j in range(0, tmp + 1, 2):
        tmp1 = 100-i - j
        for k in range(0, tmp1 + 1, 2):
            tmp2 = 100-i - j - k
            for l in range(0, tmp2 + 1, 2):
                tmp3 = 100-i - j - k - l
                for m in range(0, tmp3 + 1, 2):
                    n = 100-i - j - k-l- m
                    w = np.array([i, j, k, l, m, n])/100
                    resMat[myCount, 0:6] = w.transpose()
                    resMat[myCount, 7] = np.dot(w.transpose(), mu)
                    resMat[myCount,6] = (np.dot(np.dot(w.transpose(),Omega),w))**0.5
                    resMat[myCount, 8] = (resMat[myCount, 7] - 0.01) / resMat[myCount, 6]
                    myCount += 1
```

```python
df_stocks = pd.read_excel("Week 4 - 3stocks.xlsx")
df_stocks.head()
```

```python
mu_3stocks = df_stocks[["UU.","GSK", "RIO"]].mean(axis=0)
mu_3stocks
```

```python
Omega_3stocks = df_stocks[["UU.","GSK", "RIO"]].cov()
Omega_3stocks
```

```python
rho_3stocks = df_stocks[["UU.","GSK", "RIO"]].corr()
rho_3stocks
```

```python
onesVec = np.empty((1,3))
onesVec[:] = 1
onesVec
```

```python
A = np.dot(np.dot(onesVec,np.linalg.inv(Omega_3stocks)),mu_3stocks)[0]
A
```

```python
B = np.dot(np.dot(mu_3stocks.transpose(),np.linalg.inv(Omega_3stocks)),mu_3stocks)
B
```

```python
F = float(np.dot(np.dot(onesVec,np.linalg.inv(Omega_3stocks)),onesVec.transpose()))
F
```

```python
D = B*F-A**2
D
```

```python
alpha_0 = (np.dot((B*np.linalg.inv(Omega_3stocks)),onesVec.transpose()).reshape(3) - np.dot((A*np.linalg.inv(Omega_3stocks)),mu_3stocks.transpose()))/D
alpha_0
```

```python
alpha_1 = (np.dot((F*np.linalg.inv(Omega_3stocks)),mu_3stocks) - np.dot((A*np.linalg.inv(Omega_3stocks)),onesVec.transpose().reshape(3)))/D
alpha_1
```

```python
rbarVec = np.arange(0.6,1.51,0.01)
rbarVec
```

```python
stdVec = np.empty((len(rbarVec),1))
stdVec[:] = np.NaN
```

```python
for i in range(len(rbarVec)):
    rbar = rbarVec[i]
    w1 = alpha_0 + alpha_1*rbar
    stdVec[i] = (np.dot(np.dot(w1.transpose(),Omega_3stocks),w1))**0.5
```
